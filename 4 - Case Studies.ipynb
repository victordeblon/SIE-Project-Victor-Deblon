# Case Studies

## Analysis of precipitation measurements at surrounding stations

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

### Case study 1

```python
data = pd.read_csv('./data/Case Studies/pluvio_20191220.txt', delim_whitespace=True, skiprows=12, 
                   names=['Station', 'Year', 'Month', 'Day', 'Hour', 'Minute', 'Precipitation'], encoding='Latin')

station_names = {32:'Payerne', 749:'Romont', 850:'Oron', 14:'Grangeneuve'}
data['Station'] = data['Station'].map(station_names)
stations = data['Station'].copy().unique()

data
```

```python
# Creating subplots with specified size
fig, axs = plt.subplots(2, 2, figsize=(10, 5), sharey=True)  
axs = axs.flatten()

# Date Formatter and Locator
hour_fmt = mdates.DateFormatter('%H:%M')  # Formatter to display hours and minutes
hour_loc = mdates.HourLocator(interval=2)  # Locator to place ticks every 2 hours (adjust as needed)

for i, station in enumerate(stations):
    station_data = data[data['Station'] == station].copy()
    station_data['DateTime'] = pd.to_datetime(station_data[['Year', 'Month', 'Day', 'Hour', 'Minute']])
    axs[i].plot(station_data['DateTime'], station_data['Precipitation'], label=f'Station {station}')
    axs[i].set_title(f'{station}')
    axs[i].set_ylabel('Precipitation (mm)')
    axs[i].xaxis.set_major_locator(hour_loc)  # Set major tick locator
    axs[i].xaxis.set_major_formatter(hour_fmt)  # Set major tick formatter
    axs[i].tick_params(axis='x', rotation=45)  # Rotate x-axis labels for readability
    axs[i].grid(alpha=0.5)

plt.tight_layout()
plt.savefig('./figures/precip_case_study_1.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
